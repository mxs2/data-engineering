## Modelagem e Persistência dos Dados de Universidades

### Estrutura retornada pela API
Cada registro possui:
- `name`, `country`, `alpha_two_code`, `state-province` — campos escalares
- `domains`, `web_pages` — **listas** (um-para-muitos)

### Modelo relacional proposto

```
universities
  ├── id              INTEGER  PK AUTOINCREMENT
  ├── name            TEXT     NOT NULL
  ├── country         TEXT     NOT NULL
  ├── alpha_two_code  TEXT
  ├── state_province  TEXT
  └── extracted_at    TIMESTAMP  DEFAULT CURRENT_TIMESTAMP

university_domains
  ├── id              INTEGER  PK AUTOINCREMENT
  ├── university_id   INTEGER  FK → universities.id
  └── domain          TEXT

university_web_pages
  ├── id              INTEGER  PK AUTOINCREMENT
  ├── university_id   INTEGER  FK → universities.id
  └── web_page        TEXT
```

**Restrição de unicidade**: `UNIQUE(name, country)` em `universities` é a chave natural que evita duplicatas ao reprocessar países.

### Estratégia para múltiplos países

| Aspecto | Decisão |
|---|---|
| **Upsert** | `INSERT OR IGNORE` respeita a constraint e não sobrescreve dados existentes; use `INSERT OR REPLACE` se precisar atualizar |
| **Rastreabilidade** | `extracted_at` registra quando cada lote foi inserido |
| **Atomicidade** | Cada país é inserido dentro de uma **transação** — falha em um país não corrompe os demais |
| **Performance** | `executemany` + `commit` único por lote é muito mais rápido que commits individuais |
| **Índices** | Criar índice em `country` e `alpha_two_code` para consultas por país |

````python
uni_conn = sqlite3.connect("./data/universities.db")
cur = uni_conn.cursor()

cur.executescript("""
    CREATE TABLE IF NOT EXISTS universities (
        id              INTEGER  PRIMARY KEY AUTOINCREMENT,
        name            TEXT     NOT NULL,
        country         TEXT     NOT NULL,
        alpha_two_code  TEXT,
        state_province  TEXT,
        extracted_at    TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        UNIQUE(name, country)
    );

    CREATE TABLE IF NOT EXISTS university_domains (
        id              INTEGER  PRIMARY KEY AUTOINCREMENT,
        university_id   INTEGER  NOT NULL REFERENCES universities(id),
        domain          TEXT     NOT NULL
    );

    CREATE TABLE IF NOT EXISTS university_web_pages (
        id              INTEGER  PRIMARY KEY AUTOINCREMENT,
        university_id   INTEGER  NOT NULL REFERENCES universities(id),
        web_page        TEXT     NOT NULL
    );

    CREATE INDEX IF NOT EXISTS idx_uni_country      ON universities(country);
    CREATE INDEX IF NOT EXISTS idx_uni_alpha_two    ON universities(alpha_two_code);
    CREATE INDEX IF NOT EXISTS idx_domains_uni_id   ON university_domains(university_id);
    CREATE INDEX IF NOT EXISTS idx_webpages_uni_id  ON university_web_pages(university_id);
""")

uni_conn.commit()
print("Tabelas criadas com sucesso.")
````

````python
def persist_universities(records: list[dict], conn: sqlite3.Connection) -> int:
    """
    Insere (ou ignora duplicatas) um lote de universidades retornado pela API.
    Retorna o número de registros efetivamente inseridos.
    """
    cur = conn.cursor()
    inserted = 0

    for uni in records:
        cur.execute(
            """
            INSERT OR IGNORE INTO universities (name, country, alpha_two_code, state_province)
            VALUES (?, ?, ?, ?)
        """,
            (
                uni.get("name"),
                uni.get("country"),
                uni.get("alpha_two_code"),
                uni.get("state-province"),
            ),
        )

        if cur.rowcount == 0:
            continue

        uni_id = cur.lastrowid
        inserted += 1

        cur.executemany(
            "INSERT INTO university_domains (university_id, domain) VALUES (?, ?)",
            [(uni_id, d) for d in uni.get("domains", [])],
        )
        cur.executemany(
            "INSERT INTO university_web_pages (university_id, web_page) VALUES (?, ?)",
            [(uni_id, w) for w in uni.get("web_pages", [])],
        )

    conn.commit()
    return inserted


COUNTRIES = ["Brazil", "China", "Indonesia"]
BASE_URL = "http://universities.hipolabs.com/search?country={}"

for country in COUNTRIES:
    resp = requests.get(BASE_URL.format(country))
    resp.raise_for_status()
    data = resp.json()
    n = persist_universities(data, uni_conn)
    print(f"{country:12s} → {len(data):4d} registros na API | {n:4d} inseridos agora")
````

````python
uni_pd = lambda sql: pd.read_sql(sql, uni_conn)

# Total por país
display(uni_pd("""
    SELECT country, alpha_two_code, COUNT(*) AS total_universities
    FROM universities
    GROUP BY country
    ORDER BY total_universities DESC
"""))
````

````python
# Universidades com múltiplos domínios (JOIN com tabela filha)
display(uni_pd("""
    SELECT u.name, u.country, COUNT(d.id) AS domain_count
    FROM universities u
    JOIN university_domains d ON d.university_id = u.id
    GROUP BY u.id
    HAVING domain_count > 1
    ORDER BY domain_count DESC
    LIMIT 10
"""))
````
